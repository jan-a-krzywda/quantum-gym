```python
import importlib

importlib.reload(importlib)
importlib.reload(__import__('tuna_fid_single_job'))
from qiskit_quantuminspire.qi_provider import QIProvider
from qiskit import transpile
import pickle

from tuna_fid_single_job import build_fid_circuit_all_qubits, tau_ns_from_indices
```

```python
provider = QIProvider()
backend = provider.get_backend("Tuna-17")

taus = tau_ns_from_indices(1, 51, dt=200.0)  # 200 ns .. 10000 ns
dt = 20e-9
shots = 5000
echo = True
reset = False

qc = build_fid_circuit_all_qubits(
    backend.num_qubits,
    taus,
    dt_seconds=dt,
    echo=echo,
    OTOC = False
)
tqc = transpile(qc, backend, optimization_level=3)
job = backend.run(qc, shots=shots, memory=True)
result = job.result(timeout=6000)
mem = result.get_memory()
#mem = None
payload = {
    "memory": mem,
    "taus": list(taus),
    "shots": shots,
    "echo": echo,
    "reset": reset,
}
print(f"Collected shots: {len(mem)}")
```

```python
from pathlib import Path

data_dir = Path("data")
data_dir.mkdir(parents=True, exist_ok=True)
out_pickle = data_dir / "fid_job_memory_noreset_large_14004echo.pkl"
with out_pickle.open("wb") as f:
    pickle.dump(payload, f)
print("Wrote:", out_pickle)
print(
    "Saved metadata:",
    {
        "taus": (payload["taus"][0], payload["taus"][-1], len(payload["taus"])),
        "shots": payload["shots"],
        "echo": payload["echo"],
        "reset": payload["reset"],
    },
)
```
